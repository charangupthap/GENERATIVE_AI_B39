```python
import numpy as np

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Linear activation function (output layer)
def linear(x):
    return x

# Derivative of the linear function (for backpropagation)
def linear_derivative(x):
    return 1

# Training data (Table 1)
X_train = np.array([[0.2, 0.1],
                    [0.3, 0.2],
                    [0.4, 0.3],
                    [0.7, 0.6],
                    [0.8, 0.7],
                    [0.9, 0.8]])

y_train = np.array([[0.3441],
                    [0.3500],
                    [0.3558],
                    [0.3729],
                    [0.3785],
                    [0.3841]])

# Initialize weights and biases (random values for simplicity)
np.random.seed(42)
input_layer_neurons = 2  # Two input features (x1, x2)
hidden_layer_neurons = 4  # Number of neurons in the hidden layer
output_layer_neurons = 1  # Single output

# Weights initialization
weights_input_hidden = np.random.rand(input_layer_neurons, hidden_layer_neurons)
weights_hidden_output = np.random.rand(hidden_layer_neurons, output_layer_neurons)

# Bias initialization
bias_hidden = np.random.rand(1, hidden_layer_neurons)
bias_output = np.random.rand(1, output_layer_neurons)

# Learning rate
learning_rate = 0.1

# Train the network (Backpropagation and Gradient Descent)
epochs = 10000  # Number of iterations
for epoch in range(epochs):
    # Forward propagation
    hidden_layer_input = np.dot(X_train, weights_input_hidden) + bias_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    predicted_output = linear(output_layer_input)

    # Calculate the error (MSE)
    error = y_train - predicted_output
    mean_squared_error = np.mean(error**2)

    # Backpropagation (Gradient Descent)
    # Output layer gradients
    output_error = error
    output_delta = output_error * linear_derivative(predicted_output)

    # Hidden layer gradients
    hidden_error = output_delta.dot(weights_hidden_output.T)
    hidden_delta = hidden_error * sigmoid_derivative(hidden_layer_output)

    # Update weights and biases
    weights_hidden_output += hidden_layer_output.T.dot(output_delta) * learning_rate
    bias_output += np.sum(output_delta, axis=0, keepdims=True) * learning_rate

    weights_input_hidden += X_train.T.dot(hidden_delta) * learning_rate
    bias_hidden += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate

    # Print MSE every 1000 epochs for monitoring
    if epoch % 1000 == 0:
        print(f"Epoch {epoch} - MSE: {mean_squared_error}")

# Test the model using provided test data (Table 2)
X_test = np.array([[0.5, 0.4],
                   [0.6, 0.5]])

y_test = np.array([[0.3615],
                   [0.3672]])

# Predict on test data
hidden_layer_input_test = np.dot(X_test, weights_input_hidden) + bias_hidden
hidden_layer_output_test = sigmoid(hidden_layer_input_test)

output_layer_input_test = np.dot(hidden_layer_output_test, weights_hidden_output) + bias_output
predicted_test_output = linear(output_layer_input_test)

# Calculate MSE for test data
test_error = y_test - predicted_test_output
test_mse = np.mean(test_error**2)
print(f"Test MSE: {test_mse}")

# User Input for Prediction
def predict_user_input():
    x1 = float(input("Enter value for x1: "))
    x2 = float(input("Enter value for x2: "))
    X_user = np.array([[x1, x2]])

    hidden_layer_input_user = np.dot(X_user, weights_input_hidden) + bias_hidden
    hidden_layer_output_user = sigmoid(hidden_layer_input_user)

    output_layer_input_user = np.dot(hidden_layer_output_user, weights_hidden_output) + bias_output
    predicted_output_user = linear(output_layer_input_user)

    print(f"Predicted Output: {predicted_output_user[0][0]}")

# Predict based on user input
predict_user_input()
```
